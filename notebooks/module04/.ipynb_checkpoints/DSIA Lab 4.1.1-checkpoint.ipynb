#### 2.3 What is the simplest way to predict price

So we have information for 985 houses. What is the easiest way to predict price of a house?

#### 2. 4 Find Resudals

Find difference between actual value and predicted value. Use absolute value.

```python
# ANSWER

def find_residuals(preds, actual):
    
    abs_diff = [abs(actual[i] - preds[i]) for i in range(len(preds))]
    
    return abs_diff
```

```python
# View first 10 residuals

find_residuals(house_csv['pred_price_by_mean'].tolist(), house_csv['price'].tolist())[:10]
```

#### 2.5  Predict Using Linear Regression

[Simple Linear Regression](https://en.wikipedia.org/wiki/Simple_linear_regression#Fitting_the_regression_line)

### $$\hat{y} = \beta_0 + \beta_1 x$$

> ### $$ \beta_1 = \frac{\sum_{i=1}^n (y_i - \bar{y} ) (x_i - \bar{x} )}{\sum_{i=1}^n (x_i - \bar{x})^2} $$

and

> ### $$ \beta_0 = \bar{y} - \beta_1\bar{x} $$

Given variables:
- $\bar{y}$ : the sample mean of observed values $Y$
- $\bar{x}$ : the sample mean of observed values $X$

##### 2.5 Define X, Y

- X = Predictor Variable 
- Y = Target Variable

```python
# ANSWER

X = house_csv['sq__ft']

Y = house_csv['price']
```

##### 2.5.A Calculate beta_1

 ### $$ \beta_1 = \frac{\sum_{i=1}^n (y_i - \bar{y} ) (x_i - \bar{x} )}{\sum_{i=1}^n (x_i - \bar{x})^2} $$

```python
# ANSWER

beta_1 = ((Y - Y.mean()) * (X - X.mean())).sum() / ((X-X.mean())**2).sum()

beta_1
```

##### 2.5.B Calculate beta_0

### $$ \beta_0 = \bar{y} - \beta_1\bar{x} $$

```python
# ANSWER

beta_0 = Y.mean() - (beta_1 * X.mean())
beta_0
```

```python
def calculate_univariate_regression_coeffs(X, Y):
    
    beta_1 = ((Y - Y.mean()) * (X - X.mean())).sum() / ((X-X.mean())**2).sum()
    beta_0 = Y.mean() - (beta_1 * X.mean())
    
    coeffs = [beta_0, beta_1]
    
    return coeffs
```

```python
area_coeffs = calculate_univariate_regression_coeffs(house_csv['sq__ft'], house_csv['price'])

area_coeffs
```

##### 2.4.C Find Yhat (Predicted Value Using Beta 0 and Beta 1)

### $$\hat{y} = \beta_0 + \beta_1 x$$

```python
# ANSWER

house_csv['pred_price_linreg'] = beta_0 + (beta_1 * house_csv['sq__ft'])
```

```python
def predict_univariate_regression(X, Y):
    
    coeffs = calculate_univariate_regression_coeffs(X, Y)
    
    return coeffs[0] + (coeffs[1] * X)
```

##### 2.5.D Find Resudals

### $$ \text{residual}_i = y_i - \hat{y}_i$$

```python
# ANSWER

house_csv['residuals'] = house_csv['price'] - house_csv['pred_price_linreg']
```

##### 2.5.D Find R-Square

The coefficient $R^2$ is defined as $$ R^2 = 1 - \frac{SS_{res}}{SS_{tot}} $$, where $$ SS_{res} $$ is the residual sum of squares $$ SS_{res} = \sum_{i=1}^n \left(y_i - \hat{y}_i\right)^2 $$ and $$ SS_{res} $$ is the total sum of squares $$ SS_{tot} = \sum_{i=1}^n \left(y_i - \bar{y}\right)^2 $$

The best possible score is 1.0 and it can be negative (because the model can be arbitrarily worse).

**Write a function to calculate R- Square value.**

```python
def find_r_square(preds, actual):
    
    ss_res = ((actual - preds)**2).sum()
    ss_total = ((actual - actual.mean())**2).sum()
    
    r_sq = 1 - (ss_res / ss_total)
    
    return r_sq
```

The total sum of squares:

### $$ SS_{tot} = \sum_{i=1}^n \left(y_i - \bar{y}\right)^2 $$

The residual sum of squares:

### $$ SS_{res} = \sum_{i=1}^n \left(y_i - \hat{y}_i\right)^2 $$

$R^2$:

### $$ R^2 = 1 - \frac{SS_{res}}{SS_{tot}} $$

```python
# ANSWER

find_r_square(house_csv['pred_price_linreg'], house_csv['price'])
```

##### 2.5.G [BONUS] Draw Plot of Predicted Line

```python
def plot_univariate_regression(df, X_col, preds_col, actual_col, xlabel, ylabel, residuals=False):
    
    plt.figure(figsize=(15, 12))
    
    # Plot X vs. actual
    plt.scatter(df[X_col], df[actual_col], alpha=0.6)
    
    # Plot linear regression line
    
    plt.plot(df[X_col], # x points
             df[preds_col], # pred points
             c='red')
    
    # Plot residuals on linear regression
    if residuals == True:
        for i, row in df.iterrows():
            
            # For every row in df, plot a line between (x, yhat) and (x, y)
            plt.plot([row[X_col], row[X_col]],
                     [row[preds_col], row[actual_col]],
             c='grey',
             linestyle='dashed',
             linewidth=1,
             alpha=0.7)
        
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
```

```python
plot_univariate_regression(house_csv, 'sq__ft', 'pred_price_linreg', 'price', 
                           'Area (sq. ft)', 'Price ($)')
plt.show()
```

##### 2.5.G [BONUS] Draw Plot of Predicted Line With Resudals

```python
# ANSWER

plot_univariate_regression(house_csv, 'sq__ft', 'pred_price_linreg', 'price', 
                           'Area (sq. ft)', 'Price ($)', residuals=True)
plt.show()
```

```python
def plot_residuals(df, X, preds_col, actual_col, xlabel, ylabel="Residual"):
    
    residuals = df[actual_col] - df[preds_col]
    
    plt.figure(figsize=(15, 12))

    plt.scatter(df[actual_col], residuals, alpha=0.6)
    plt.axhline(0, linestyle='dashed', c='red', alpha=0.7)

    plt.xlabel(xlabel)
    plt.ylabel("Residual")
```

```python
plot_residuals(house_csv, 'sq__ft', 'pred_price_linreg', 'price', 'Price ($)')
plt.show()
```

```python
# The data has a lot of variability in price when area is 0. Will predictions improve if we omit this data?

house_filtered = house_csv[house_csv['sq__ft'] > 0]

house_filtered_coeffs = calculate_univariate_regression_coeffs(house_filtered['sq__ft'], 
                                                               house_filtered['price'])

house_filtered['pred_price_linreg'] = predict_univariate_regression(house_filtered_coeffs,
                                                                    house_filtered['sq__ft'])

house_filtered['residuals'] = house_filtered['price'] - house_filtered['pred_price_linreg']

print("Regression R2: {}".format(find_r_square(house_filtered['pred_price_linreg'], house_filtered['price'])))
```

```python
# How many records are left if we omit houses with sq. ft == 0?

house_filtered.shape
```

```python
# Visualise linear regression

plot_univariate_regression(house_filtered, 'sq__ft', 'pred_price_linreg', 'price', 
                           'Area (sq. ft)', 'Price ($)', residuals=True)
plt.show()
```

```python
# Plot residuals

plot_residuals(house_filtered, 'sq__ft', 'pred_price_linreg', 'price', 'Price ($)')
plt.show()
```

### 3. Predict House Price

- city:	SACRAMENTO
- zip:	95833
- state:	CA
- beds:	3
- baths:	2
- sq__ft:	1204
- type:	Residential

**Predict the price of this house.**

```python
# ANSWER

pred_price = predict_univariate_regression(area_coeffs, 1204)
pred_price2 = predict_univariate_regression(house_filtered_coeffs, 1204)

print("Predicted price for a house with {} sq. ft is ${}.".format(1204, round(pred_price, 2)))
print("Predicted price for a house with {} sq. ft is ${} when sq. ft = 0 omitted".format(1204, round(pred_price2, 2)))
```

```python
# What is the actual price of this house?

actual_price = house_csv.loc[(house_csv['sq__ft'] == 1204) &
                             (house_csv['street'] == '1140 EDMONTON DR'), 'price'].values 

actual_price
```

```python
# Calculate residual

print("Residual using original model: ", actual_price - pred_price)
print("Residual using modified model: ", actual_price - pred_price2)
```
