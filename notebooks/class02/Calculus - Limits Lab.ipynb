# Calculus - Limits

```python
import numpy as np
import matplotlib.pyplot as plt
# get matplot to show charts inline
%matplotlib inline
```

```python
# define a function, say f(x) = 1/x
def f(x):
    return 1/x
```

```python
# calculate value of f(x) for some values
print('f(4) = ',f(4))
print('f(5) = ',f(5))
print('f(6) = ',f(6))
```

```python
# calculate value of f(x) for 0
try:
    print('f(0) = ',f(0))
except ZeroDivisionError:
    print("Can't divide by zero!")    
```

```python
# Calculate f(x) for a very small value of x

print('f(0.2) = ',f(0.2))
```

```python
# plot f(x) from a small number to a large number
x = np.arange(0.1,100,1)
y = f(x)

plt.figure(figsize=(8, 8))
plt.grid()
plt.plot(x, y)
plt.title("Plot with both axes at same scale")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```

```python
# plot f(x) from a small number to a large number with logarithmic scale
x = np.arange(0.1,100,1)
y = f(x)

plt.figure(figsize=(8, 8))
plt.grid()
plt.semilogy(x, y)
plt.title("Plot with y-axis at log scale")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```

```python
# Add another function

def g(x):
  return np.log(x) + 1
```

```python
# Test function

print("g(1) = ", g(1))
print("g(6) = ", g(6))
print("g(0.1) = ", g(0.1))
```

```python
# plot f(x) from a small number to a large number
x = np.arange(0.1,100,1)
y = g(x)

plt.figure(figsize=(8, 8))
plt.grid()
plt.plot(x, y, color="green")
plt.title("Plot with both axes at same scale", fontweight="bold")
plt.xlabel("x",  fontsize = 12,  fontweight="bold")
plt.ylabel("g(x)", fontsize = 12,  fontweight="bold")
plt.show()
```
