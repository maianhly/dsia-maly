# Statistics - part 1

```python
import numpy as np
import matplotlib.pyplot as plt
# get matplot to show charts inline
%matplotlib inline
```

```python
# plot frequency of some observations using pie chart
labels = ['Developers', 'Data Engineers', 'Data Scientists', 'Business Analysts']
frequency = [15, 32, 12, 7]

fig1, ax1 = plt.subplots()
explode = (0, 0, 0.2, 0)  # only "explode" the 3rd slice (i.e. 'Data Scientist')
ax1.pie(frequency, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# plot above as bar graph

plt.figure(figsize=(12, 8))
plt.gca().yaxis.grid(True)
barlist = plt.bar(np.arange(len(labels)), frequency, align='center')

for i, color in enumerate(['red', 'blue', 'green', 'orange']):
    barlist[i].set_color(color)

plt.ylabel("Number of people", fontweight="bold")
plt.xticks(np.arange(len(labels)), labels, fontweight="bold")
plt.show()
```

```python
norm_freq = [x - frequency[2] for x in frequency]

plt.figure(figsize=(12, 8))
plt.gca().yaxis.grid(True)
barlist = plt.bar(np.arange(len(labels)), norm_freq)
barlist[2].set_color('r')
plt.ylabel("Number of people")
plt.xticks(np.arange(len(labels)), labels, rotation=60)
plt.show()
```

```python
# plot frequency of some observations using bar chart
objects = ('Python', 'C++', 'Java', 'Perl', 'Scala', 'Lisp')
y_pos = np.arange(len(objects))
performance = [10,8,6,4,2,1]
 
# Create a bar plot

plt.figure(figsize=(10, 6))
plt.bar(y_pos, performance, align = 'center', color=['red', 'blue', 'green', 'orange'])
plt.xticks(y_pos, objects)
plt.ylabel('Usage')
plt.xlabel('Language')
plt.title('Programming language usage')
 
plt.show()
```
