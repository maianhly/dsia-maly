# Calculus - Integral

```python
import numpy as np
import matplotlib.pyplot as plt
# get matplot to show charts inline
%matplotlib inline
```

```python
# define a function such as f(x) = x**2
def f(x):
  '''
  Return x**2
  '''
  return x**2
```

```python
# plot f(x)
x = np.arange(-10, 11, 1)
y = f(x)

plt.figure(figsize=(8, 8))
plt.plot(x, y)
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')

plt.show();
```

```python
# test the function you defined above by simply print its value for some numbers
print('f(2) = ', f(2))
print('f(5) = ', f(5))
print('f(-3) = ', f(-3))
```

```python
# define integral of f(x)
# integral of a function is the area under the curve of this function
# it can be approximated by plotting lots of retangles and add their areas
# the more reactangles you draw, the better is the approximation
# but why do it by hand where python can do a much better job?!
def integral(startX, endX, numberOfRectangles):
    width = (float(endX) - float(startX))/ numberOfRectangles
    totalArea = 0
    for i in range(numberOfRectangles):
        height = f(startX + i*width)
        area = width * height
        totalArea += area
    return totalArea
```

```python
# calculate integral of f(x) for some startX, endX and a number of rectangles
print('integral 2 to 4, 10 rectangles drawn = ',integral(2.0,4.0,10))
print('integral 2 to 4, 1000 rectangles drawn = ',integral(2.0,4.0,1000))
print('-----')
print('integral -1 to 2, 10 rectangles drawn = ',integral(-1.0,2.0,10))
print('integral -1 to 2, 1000 rectangles drawn = ',integral(-1.0,2.0,1000))
```

```python
import time
```

```python
# How long does the integral function take to draw many rectangles?

num_rect = np.arange(5, 1000000, 10)
area_under_curve = []
times = []

for i in num_rect:
  start = time.time()
  area = integral(2, 4, i)
  end = time.time()
  exe_time = end - start
  area_under_curve.append(area)
  times.append(exe_time)
  
plt.figure(figsize=(10, 6))
plt.plot(num_rect, times)
plt.xlabel("Number of rectangles drawn")
plt.ylabel("Time executed (sec)")
plt.show()
  
```
