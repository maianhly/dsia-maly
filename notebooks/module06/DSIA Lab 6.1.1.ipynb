### Australian athletes data set

##### Description
These data were collected in a study of how data on various characteristics of the bloood varied with sport body size and sex of the athlete.

Usage
data(ais)
Format
A data frame with 202 observations on the following 13 variables.

`rcc` - red blood cell count, in per liter

`wcc` - white blood cell count, in per liter

`hc` - hematocrit, percent

`hg` - hemaglobin concentration, in g per decaliter

`ferr` - plasma ferritins, ng

`bmi` - Body mass index, kg

`ssf` - sum of skin folds

`pcBfat` - percent Body fat

`lbm` - lean body mass, kg

`ht` - height, cm

`wt` - weight, kg

`sex` - a factor with levels f m

`sport` - a factor with levels B_Ball Field Gym Netball Row Swim T_400m T_Sprnt Tennis W_Polo

Details
Do blood hemoglobin concentrations of athletes in endurance-related events differ from those in power-related events?

Source
These data were the basis for the analyses that are reported in Telford and Cunningham (1991).

References
Telford, R.D. and Cunningham, R.B. 1991. Sex, sport and body-size dependency of hematology in highly trained athletes. Medicine and Science in Sports and Exercise 23: 788-794.

**References**

[Data](http://vincentarelbundock.github.io/Rdatasets/csv/DAAG/ais.csv)

[Data Dictionary](http://vincentarelbundock.github.io/Rdatasets/doc/DAAG/ais.html)

### 1. Import the data

```python
csv_file = pd.read_csv('../../data/ais.csv', index_col=0)

csv_file.head()
```

### 2. Do some EDA of relationships between features.

```python
print("Data has {} observations and {} features".format(csv_file.shape[0], csv_file.shape[1]))
```

```python
csv_file.describe()
```

#### Find correlations

```python
# Get pairplot of numeric features

num_cols = csv_file.select_dtypes(include="number").columns.tolist()

sns.pairplot(csv_file[num_cols])
plt.show()
```

```python
# Get heatmap of correlations

plt.figure(figsize=(10, 10))
sns.heatmap(csv_file[num_cols].corr(), annot=True)
plt.show()
```

- `rcc`, `wcc`, `hc`, `ferr`, `ssf` have skewed normal distributions, where smaller values are more common

- `rcc`, `hc` and `ferr` have strong positive correlation

#### Find distributions of numeric features by sex

```python
# Color distribution of numeric features by sex

fig = plt.figure(figsize=(15, 50))

for i, col in enumerate(num_cols):
    ax = fig.add_subplot(len(num_cols), 3, 1+i)
    n, bins, patches = ax.hist(csv_file[csv_file['sex'] == 'm'][col], label='Male', color='darkblue', alpha=0.6)
    ax.hist(csv_file[csv_file['sex'] == 'f'][col], bins=bins, label='Female', color='darkred', alpha=0.6)
    ax.set_title("Distribution of {}".format(col), size=12)
    ax.set_xlabel("", labelpad=10)
    
    if i == 0:
        ax.legend(title='Gender', loc='upper right')    
        
plt.show()    
```

#### Find outliers

```python
# Find quantiles

outlier_quantiles = csv_file[num_cols].quantile([0.01, 0.97])
outlier_quantiles
```

```python
# Remove observations 3 std from the mean

no_outliers = csv_file[num_cols].mask((csv_file[num_cols] - csv_file[num_cols].mean()).abs() > 3 * csv_file[num_cols].std()).dropna()

no_outliers.shape
```

```python
no_outliers.head()
```

### 3. Prepare the data for clustering

1. Remove the `sex`, `sport` column. We will see if the clusters from K-Means end up like the actual sex.
2. Put the features on the same scale.

```python
# ANSWER

scaler = StandardScaler()

scaledX = scaler.fit_transform(no_outliers)
```

### 4. Clustering with K-Means

- Cluster the data to our our target groups.
    - Use sklearn.cluster.KMeans
- We know that there are 2 actual classes. Lets try using 2, 4. Run the clustering algorithm with each of those k values.

```python
# ANSWER

# instantiate KMeans with different k

kmeans_2 = KMeans(n_clusters=2, random_state=12)
kmeans_4 = KMeans(n_clusters=4, random_state=12)
```

### 5. Get the labels for the models

```python
# ANSWER

no_outliers['preds_k2'] = kmeans_2.fit_predict(scaledX)
no_outliers['preds_k4'] = kmeans_4.fit_predict(scaledX)
```

```python
no_outliers.head()
```

### 6. Elbow Method For Optimal k

For each k value, we will initialise k-means and use the inertia attribute to identify the sum of squared distances of samples to the nearest cluster centre. As k increases, the sum of squared distance tends to zero. 

```python
# ANSWER

# Plot inertia for every k until k = 30

inertia = []
K = 30

for i in range(K):
    
    kmeans = KMeans(n_clusters=i+1, random_state=12)
    kmeans.fit(scaledX)
    inertia.append(kmeans.inertia_)
    
plt.figure(figsize=(15, 6))
plt.plot(np.arange(K)+1, inertia, marker="o")
plt.ylabel('Inertia')
plt.xlabel('K clusters')

plt.show()
```

Above is a plot of sum of squared distances for k in the range specified above. If the plot looks like an arm, then the elbow on the arm is optimal k.

### 7. Plot `ht` vs. `wt` using the clusters from each model.

- Plot ht, wt using hue sex
- Plot ht, wt using hue label

Each time, compare to the actual species classification.

```python
# ANSWER

# Join sex, sport to no_outliers df

all_f = no_outliers.join(csv_file[['sex', 'sport']], )

# Plot ht, wt by sex and k2 prediction

fig = plt.figure(figsize=(15, 15))

ax1 = fig.add_subplot(211)
sns.scatterplot(x='ht', y='wt', hue='sex', data=all_f, ax=ax1, alpha=0.7, s=100)

ax2 = fig.add_subplot(212)
sns.scatterplot(x='ht', y='wt', hue='preds_k2', data=all_f, ax=ax2, alpha=0.7, s=100)

plt.show()
```

```python
# Plot pcBfat, bmi by sex and k2 prediction

fig = plt.figure(figsize=(15, 15))

ax1 = fig.add_subplot(211)
sns.scatterplot(x='pcBfat', y='bmi', hue='sex', data=all_f, ax=ax1, alpha=0.7, s=100)

ax2 = fig.add_subplot(212)
sns.scatterplot(x='pcBfat', y='bmi', hue='preds_k2', data=all_f, ax=ax2, alpha=0.7, s=100)

plt.show()
```

```python
# Compare k2 predictions to gender label

f0 = all_f[(all_f['sex'] == 'f') & (all_f['preds_k2'] == 0)].shape[0]
f1 = all_f[(all_f['sex'] == 'f') & (all_f['preds_k2'] == 1)].shape[0]
m0 = all_f[(all_f['sex'] == 'm') & (all_f['preds_k2'] == 0)].shape[0]
m1 = all_f[(all_f['sex'] == 'm') & (all_f['preds_k2'] == 1)].shape[0]

print("{}% females grouped in cluster 0".format(round(f0 / all_f[all_f['sex'] == 'f'].shape[0] * 100, 2)))
print("{}% females grouped in cluster 1".format(round(f1 / all_f[all_f['sex'] == 'f'].shape[0] * 100, 2)))
print("\n")
print("{}% males grouped in cluster 0".format(round(m0 / all_f[all_f['sex'] == 'm'].shape[0] * 100, 2)))
print("{}% males grouped in cluster 1".format(round(m1 / all_f[all_f['sex'] == 'm'].shape[0] * 100, 2)))
```

When K = 2, the majority of females are grouped in cluster 0, and the majority of males are grouped in cluster 1.
